# Data cleaning & Data merged

## Set up  and load  packages
  

```python
import pandas as pd
import json
import simplejson as json
import numpy as np
```

### For business dataset, we load json file and transfer to a dataframe.

```python

users = []
with open('./yelp_dataset/yelp_business.json') as fl:
    for i, line in enumerate(fl):
        users.append(json.loads(line))
df = pd.DataFrame(users)
df.head()
```

### Check the shape of business dataset

```python
df.shape
```

### Check the  statistical information of numeric arrtibutes in business dataset

```python
df.describe()
```

### Check the data type of each variable in business dataset

```python
df.info()
```

### Get the summary of NaN value in business dataset

```python
df.isnull().sum()
```

### Get the cleaned_df by drop off all NaN value 

```python
cleaned_df = df.dropna()
```

### Get the subset of business dataset 
*  Filter the DataFrame to include only rows where the "categories" column contains "Restaurants"

```python

filtered_df = cleaned_df.loc[cleaned_df['categories'].str.contains('Restaurants')]
```

```python
filtered_df.head()
```

* Drop the "arritibutes" and "hours" columns from the DataFrame because we don't need to analysis this two variables.

```python

new_df = filtered_df.drop(['attributes', 'hours'], axis=1)
```

```python
new_df.head()
```

### Load yelp_tip.json file and read as a dataframe

```python
# Read a JSON file containing multiple objects separated by newlines
tip_df = pd.read_json('./yelp_dataset/yelp_tip.json', lines=True)

# Print the DataFrame
tip_df.head()
```

### Get the shape of tip dataset

```python
tip_df.shape
```

### Get the datatype of each variable in tip dataset

```python
tip_df.info()
```

### Get the summary of NaN value in tip dataset

```python
tip_df.isnull().sum()
```

### Merge cleaned business dataset and tip dataset based on business_id 

```python
# Merge df1 and df2 on the 'business_id' column, keeping only the rows with matching IDs
merged_df = pd.merge(new_df, tip_df, on='business_id', how='inner')

# Print the merged DataFrame
merged_df.head()
```

### Save merged dataset and name it as Output1.csv

```python
merged_df.to_csv('output1.csv', index=False)
```
