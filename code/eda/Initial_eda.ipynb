### Read the datasets

```python
import pandas as pd
df = pd.read_json("../..data/raw_data/yelp_dataset/yelp_academic_dataset_review.json",lines = True, chunksize=100000)
df1 = pd.DataFrame()
for chunk in df:
    df1 = pd.concat([df1, chunk])
df1.head()
```

```python
df2 = pd.read_csv("../..data/clean_data/output1.csv")
df2.head()
```

```python
business = df2[["business_id","name","address","city","state","postal_code","latitude","longitude","stars","review_count","is_open","categories","compliment_count"]]
business = business.drop_duplicates()
# business = business[(business.is_open == 1)]
print(business.head())
len(business)
```

```python
tips = df2
print(tips.head())
len(tips)
```

```python
reviews = business.merge(df1,left_on="business_id", right_on = "business_id", how="inner")
print(reviews.head())
len(reviews)
```

```python
# Write the datasets
tips.to_csv("tips.csv", sep='\t')
reviews.to_csv("reviews.csv", sep='\t')
```

### Explore the business data sort by states

```python
import matplotlib.pyplot as plt
import numpy as np
business.state.groupby(business.state).count().sort_values()[::-1][:35].plot(kind="bar")
```

Here, we can see from the above plot that PA, FL and MO states are the top three restaurants in terms of quantity. Then I focus on FL to explore more about the distribution of cities.

```python
business_FL = business[(business.state == "FL")]
business_FL.city.groupby(business.city).count().sort_values()[::-1][:35].plot(kind="bar")
```

As we can see, tampa has the most restaurants in FL.
